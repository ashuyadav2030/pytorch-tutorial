Example 3

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
print(x)
print(y)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0 , requires_grad=True)
```

```python
print(w)
print(b)
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = torch.binary_cross_entropy_with_logits(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

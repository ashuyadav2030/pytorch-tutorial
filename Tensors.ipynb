GPU IMplemtation

```python
import torch
```

```python
if torch.cuda.is_available():
  print("wow, i am using gpu")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
import time
a = torch.rand(10000,10000)
b = torch.rand(10000, 10000)

start_cpu = time.time()
c = torch.matmul(a,b)
end_cpu = time.time()

differnece_cpu = end_cpu - start_cpu
print(differnece_cpu)

a = a.to(device)
b = b.to(device)
start_gpu = time.time()
c = torch.matmul(a,b)
end_gpu = time.time()

differnece_gpu = end_gpu - start_gpu
print(differnece_gpu)
```

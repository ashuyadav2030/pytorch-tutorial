```python
# -*- coding: utf-8 -*-
"""Untitled4.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1fRp9RU51smTckIlMCO8yTlk1jMIhtH7U
"""

import torch
print(torch.__version__)

from sklearn.datasets import make_classification

# Generate synthetic data
x, y = make_classification(n_samples=10, n_features=2, n_informative=2, n_redundant=0, random_state=1)

# Convert to PyTorch tensors
X = torch.from_numpy(x.astype('float32'))
Y = torch.from_numpy(y.astype('float32'))

from torch.utils.data import Dataset, DataLoader

# Custom Dataset class
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, index):
        return self.features[index], self.labels[index]

# Create dataset and dataloader
dataset = CustomDataset(X, Y)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

# Print batches
for batch_features, batch_labels in dataloader:
    print(batch_features)
    print(batch_labels)

"""Practice Example"""

import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("BreastCancer.csv")

# Drop 'Id' column
drop_id = data.drop(['Id'], axis=1)

# Fill missing values
data_fillna = drop_id.fillna(0)

# Split features and labels
x_numpy = data_fillna.drop(['Class'], axis=1).values.astype('float32')
y_numpy = data_fillna['Class'].values.astype('float32')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(x_numpy, y_numpy, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train).astype('float32')
X_test = scaler.transform(X_test).astype('float32')

# Convert to PyTorch tensors
X_train = torch.from_numpy(X_train)
X_test = torch.from_numpy(X_test)
Y_train = torch.from_numpy(Y_train)
Y_test = torch.from_numpy(Y_test)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)

# Custom Dataset class for train and test data
train_dataset = CustomDataset(X_train, Y_train)
test_dataset = CustomDataset(X_test, Y_test)

# DataLoaders for train and test data
train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=2, shuffle=True)

import torch.nn as nn

# Define the neural network
class mynn(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 5),
            nn.ReLU(),
            nn.Linear(5, 1),
            nn.Sigmoid()
        )

    def forward(self, features):
        out = self.network(features)
        return out

# Initialize model, optimizer, and loss function
model = mynn(X_train.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr=0.000002)
loss_function = nn.BCELoss()

# Training loop
epochs = 25
for epoch in range(epochs):
    for batch_features, batch_labels in train_dataloader:
        # Ensure data types are correct and labels have the correct shape
        y_pred = model(batch_features.float())
        loss = loss_function(y_pred, batch_labels.float().unsqueeze(1))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```
